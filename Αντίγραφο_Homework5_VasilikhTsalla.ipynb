## Και από εδώ και πέρα ξεκινούν οι δικές σας 10 ερωτήσεις στα δεδομένα! Φτιάξτε γραφήματα και βρείτε π.χ. σε ποιες βιομηχανίες συμβαίνουν τα περισσότερα ατυχήματα κλπ. Καλή επιτυχία!

```python
accidents.dtypes
```

1. Ποιες είναι οι Top 10 Πολιτείες που συμβαίνουν ατυχήματα;

```python
accidents.State.value_counts().head(10).sort_values().plot(kind= 'barh')
```

```python
accidents.State.value_counts().head(10).sort_values()
```

2. Ποιες είναι οι top 5 πόλεις που γίνονται τα περισσότερα ατυχήματα;




```python
accidents['City'].value_counts().head(5).sort_values().plot(kind='barh')
```

```python
accidents['City'].value_counts().head(5).sort_values()
```

3. Ποιές ήταν οι top 10 πόλεις όπου έγιναν οι περισσότεροι ακρωτηριασμοί;



```python
accidents.groupby('City')['Amputation'].value_counts().head(10).sort_values().plot(kind= 'barh')
```

```python
accidents.groupby('City')['Amputation'].value_counts().head(10).sort_values()
```

4. Ποια top 10 πιο συχνά είδη ατυχήματος;


```python
accidents['NatureTitle'].value_counts().head(10).sort_values().plot(kind='barh')
```

```python
accidents['NatureTitle'].value_counts().head(10).sort_values()
```

5. Ποιες είναι οι top 10 αιτίες για τα ατυχήματα;

```python
accidents['SourceTitle'].value_counts().head(10).sort_values().plot(kind='barh')
```

```python
accidents['SourceTitle'].value_counts().head(10).sort_values()
```
